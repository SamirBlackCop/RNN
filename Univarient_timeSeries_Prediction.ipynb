```python
import numpy as np
from tensorflow.keras.models  import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, Flatten
```

```python
#prepareing independent and dependent featuures

import numpy

def prepare_data(timeseries_data, n_steps):
    X, y = [], []
    for i in range(len(timeseries_data)):
        end_ix = i+ n_steps
        if end_ix >len(timeseries_data)-1:
            break
                 
        seq_x, seq_y = timeseries_data[i:end_ix], timeseries_data[end_ix]
        X.append(seq_x)
        y.append(seq_y)
        
    return  numpy.array(X), numpy.array(y)
```

```python
# define The input Sequence
timeseries_data = [110,125, 133,146,158,172,187,196, 210]
# Choose a no of Time_Steps
n_steps = 3
# Split  into sample
X, y = prepare_data(timeseries_data, n_steps)
```

```python
# reshape input to be 3D [samples, time_steps, features] which is required for LSTM
n_features = 1
X = X.reshape(X.shape[0], X.shape[1], n_features)
```

### Building LSTM model


```python
# define model
model = Sequential()
model.add(LSTM(50, return_sequences=True, activation = 'relu', input_shape = (n_steps,n_features)))   #  time_steps, features as input shape   50=hidden node
model.add(LSTM(50, return_sequences=True, activation = 'relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer = 'adam')
```

```python
model.fit(X, y, verbose=2)
```

#### Predicting For the next 10 data

```python
# Demonstrate predicting for next 10 days

from numpy import array

x_input = array([187, 196, 210])
temp_input = list(x_input)
lst_output = []
i=0
while(i<10):
    if(len(temp_input)>3):
        x_input = array(temp_input[1:])
        print("{} day input {}".format(i, x_input))
        x_input = x_input.reshape(1, n_steps, n_features)
        yhat = model.predict(x_input, verbose=0)
        print("{} day output {}".format(i, yhat))
        temp_input.append(yhat[0][0])
        temp_input = temp_input[1:]
        lst_output.append(yhat[0][0])
        i=i+1
        
    else:
        x_input = x_input.reshape(1, n_steps, n_features)
        yhat = model.predict(x_input, verbose=0)
        print( yhat[0])
        temp_input.append(yhat[0][0])
        lst_output.append(yhat[0][0])
        i=i+1
        
print(lst_output)
    
```
